#### Scrapping data from multiple seasons and multiple teams using a loop

```python
years = list(range(2022,2020, -1))
```

```python
years
```

```python
#intialise a list called all matches

all_matches = []
```

```python
# use our intial url
standings_url = 'https://fbref.com/en/comps/9/11160/2021-2022-Premier-League-Stats'
```

```python
import time
for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text)
    standings_table = soup.select('table.stats_table')[0]

    links = [l.get("href") for l in standings_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]
    
    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"
    
    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")
        data = requests.get(team_url)
        matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
        soup = BeautifulSoup(data.text)
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        data = requests.get(f"https://fbref.com{links[0]}")
        shooting = pd.read_html(data.text, match="Shooting")[0]
        shooting.columns = shooting.columns.droplevel()
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError:
            continue
        team_data = team_data[team_data["Comp"] == "Premier League"]
        
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(4)
```

```python
match_df = pd.concat(all_matches)
```

#### create simple machine learning algorithms

```python
#encode the venue column. To calculate a home field advantage predictor

match_df['venue_code'] = match_df['venue'].astype('category').cat.codes
```

```python
#since machine learning does not operate with string values we have to convert them into
#categorical values with numerical encoding. We do the same with the opponents

match_df['opp_code'] = match_df['opponent'].astype('category').cat.codes
```

```python
#we need to create an hour column by removing the minute and the column so thst the hour remains

match_df['hour'] = match_df['time'].str.replace(':.+', '', regex=True).astype('int')
```

```python
#start to train our initial ML algo
#We determine our variable to be predicted, here it is a Win

match_df['target'] = (match_df['result'] == 'W').astype('int')
```

```python
#import a randomforestclassifier, this picks up non linearities in the data eg opp code in not a linear category

from sklearn.ensemble import RandomForestClassifier
```

```python
#intialise the class
#passin some paramaters (n_estimators - number of decision trees we want to train) the higher more time more acuracy
#min_samples_split - Number of samples we want to have in a leaf of the  D.tree before we split the node.
#The higher the less likely we are to over fit but the lower the accuracy
#random state - How many different results we get as we run


rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1) 
```

```python
#Generate predictions passs in test predictors

preds = rf.predict(test[predictors])
```

```python
#Determine the accuracy of the model
#accuracy score will say the percentage of the times you were correct (sum predicted win accuracy and lose accuracy)

from sklearn.metrics import accuracy_score
```

#### Improving precision using rolling averages

```python
# Create more preditors to improve acuracy by spliting match_df by teams so that we compute rolling averages

grouped_matches = match_df.groupby('team')
```

```python
group = grouped_matches.get_group('Manchester City').sort_values('date')
```

```python
#matchweek 4 example

group.head()
```

```python
#def a fnt take group, a set of cols we want to compute rolling aves for and new cols that we want to assign the aves to

def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date') # we want to look at last x matches so we sort by date
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)#drop missing values
    return group
```

```python
#the columns we want to compute moving averages for

cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f'{c}_rolling' for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group,cols,new_cols) #now we have added extra cols that have info on what happened in the previous 3 mtches

#we can pass this to our ML algo to improve it
```

```python
matches_rolling = match_df.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling
```

```python
#we drop the multi level index on first column

matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
 #our index does not coinside with the rows so we rearrange the index column to create unique values for each index
    
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```
